```
import numpy as np
import matplotlib.pyplot as plt
```

```
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
## for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```
t = np.loadtxt('tout.txt')
x_xhat_0 = np.loadtxt('x_xhat_0.txt', delimiter=',')
yrs = []
for i in range(1,8):
  yrs.append(np.loadtxt('y_r_'+str(i)+'.txt', delimiter=','))
```

```
x_legends = [r'$x_1$', r'$x_2$', r'$\hat{x}_1$', r'$\hat{x}_2$']
```

```
fig, ax = plt.subplots(1,1,figsize=(10,6))
for i in range(0, x_xhat_0.shape[1]):
  ax.plot(t[:301], x_xhat_0[:301,i], label=x_legends[i],
          linewidth=2)
plt.xlabel(r'$t$', fontsize=24, labelpad=12)
ax.tick_params(axis='both', labelsize=20)
plt.legend(fontsize=20)
plt.grid()
plt.savefig('ControlOpt_T2_LQR.pdf', bbox_inches = 'tight',
    pad_inches = 0)
plt.show()
plt.close()
```

```
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(t[:301], yrs[0][:301,0], label=r'$r$', linewidth=2)
ax.plot(t[:301], yrs[0][:301,1], label=r'$y$', linewidth=2)
plt.xlabel(r'$t$', fontsize=24, labelpad=12)
ax.tick_params(axis='both', labelsize=20)
plt.legend(fontsize=20)
plt.grid()
plt.savefig('ControlOpt_T2_LQR_ref.pdf', bbox_inches = 'tight',
    pad_inches = 0)
plt.show()
plt.close()
```
